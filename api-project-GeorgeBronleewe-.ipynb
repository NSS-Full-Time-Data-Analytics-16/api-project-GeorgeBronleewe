```python
import requests
import pandas as pd
```

1

```python
base_url = "http://api.worldbank.org/v2/country/all/indicator/NY.GDP.PCAP.PP.KD"
gdp_data = []
page = 1

while True:
    response = requests.get(base_url, params={
        "format": "json",
        "page": page
    })

    
    print(f"Status Code for Page {page}: {response.status_code}")
    
    data = response.json()
    
    if len(data) < 2:
        break

    gdp_data.extend(data[1])
    
    if page >= data[0]["pages"]:
        break
    else:
        page += 1

# Convert to DataFrame
gdp_df = pd.DataFrame(gdp_data)

# Print results
print(f"Retrieved {len(gdp_df)} rows.")
gdp_df.head()
```

2

```python
base_url = "http://api.worldbank.org/v2/country/all/indicator/SP.DYN.LE00.IN"
life_data = []
page = 1

while True:
    response = requests.get(base_url, params={
        "format": "json",
        "page": page
    })

    print(f"Status Code for Page {page}: {response.status_code}")
    
    data = response.json()
    
    if len(data) < 2:
        break

    life_data.extend(data[1])
    
    if page >= data[0]["pages"]:
        break
    else:
        page += 1

# Convert to DataFrame
life_df = pd.DataFrame(life_data)

# Show result
print(f"Retrieved {len(life_df)} rows.")
life_df.head()
```

3

```python
gdp_df_clean = gdp_df[["country", "countryiso3code", "date", "value"]].copy()
gdp_df_clean["country"] = gdp_df_clean["country"].apply(lambda x: x["value"] if isinstance(x, dict) else None)
gdp_df_clean = gdp_df_clean.rename(columns={
    "countryiso3code": "iso_code",
    "date": "year",
    "value": "gdp_per_capita"
})

# Clean Life Expectancy data
life_df_clean = life_df[["country", "countryiso3code", "date", "value"]].copy()
life_df_clean["country"] = life_df_clean["country"].apply(lambda x: x["value"] if isinstance(x, dict) else None)
life_df_clean = life_df_clean.rename(columns={
    "countryiso3code": "iso_code",
    "date": "year",
    "value": "life_expectancy"
})
```

```python
merged_df = pd.merge(gdp_df_clean, life_df_clean, on=["iso_code", "year", "country"], how="inner")
merged_df.head()
```

```python
country_url = "http://api.worldbank.org/v2/country"
country_data = []
page = 1

while True:
    response = requests.get(country_url, params={
        "format": "json",
        "page": page
    })
    
    data = response.json()
    
    if len(data) < 2:
        break

    country_data.extend(data[1])
    
    if page >= data[0]["pages"]:
        break
    else:
        page += 1

country_df = pd.DataFrame(country_data)

country_df_clean = country_df[[
    "id", "name", "region", "incomeLevel", "capitalCity", "iso2Code"
]].copy()

country_df_clean["region"] = country_df_clean["region"].apply(lambda x: x["value"] if isinstance(x, dict) else None)
country_df_clean["income_level"] = country_df_clean["incomeLevel"].apply(lambda x: x["value"] if isinstance(x, dict) else None)
country_df_clean = country_df_clean.rename(columns={
    "id": "iso_code",
    "name": "country_name",
    "capitalCity": "capital_city"
})

country_df_clean = country_df_clean.drop(columns=["incomeLevel"])  

country_df_clean.head()
```

```python
final_df = pd.merge(merged_df, country_df_clean, on="iso_code", how="left")

final_df = final_df[final_df["region"] != "Aggregates"]

print(f"Final dataset contains {len(final_df)} rows after dropping aggregates.")
final_df.head()
```
